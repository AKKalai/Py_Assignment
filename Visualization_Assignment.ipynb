#### In this assignment students have to transform iris data into 3 dimensions      and plot a 3d chart with transformed dimensions and colour each data point with specific class

```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
df
```

```python
features = ['sepal length', 'sepal width', 'petal length', 'petal width']
x = df.loc[:, features].values
y = df.loc[:,['target']].values
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2', 'principal component 3'])
```

```python
df_new = pd.concat([principalDf, df[['target']]], axis = 1)
df_new
```

```python
x_value = df_new['principal component 1']
y_value = df_new['principal component 2']
z_value = df_new['principal component 3']
```

```python
fig = plt.figure(figsize = (9,9))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_zlabel('Principal Component 3', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)

targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df_new['target'] == target
    ax.scatter(df_new.loc[indicesToKeep, 'principal component 1']
               , df_new.loc[indicesToKeep, 'principal component 2']
               , df_new.loc[indicesToKeep, 'principal component 3']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
plt.show()
```
